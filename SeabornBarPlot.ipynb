```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Write a program to draw bar plot of sex against survived for a dataset given in
#the url https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv 
titanic_dataset = sns.load_dataset("titanic")
sns.barplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

```python
#2. Write a program to draw a barplot to show count for a deck for a dataset given in
#the url https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv 
titanic_dataset = sns.load_dataset("titanic")
sns.countplot(x='deck', data = titanic_dataset)
plt.show()
```
