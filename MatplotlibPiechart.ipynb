```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
#1. Write a Python programming to create a pie chart of the popularity of programming Languages. 
#Sample data: 
#Programming languages: Java, Python, PHP, JavaScript, C#, C++ 
#Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
plt.axis("equal")
plt.pie(popularity, labels = languages, radius=1.5, autopct = "%0.1f%%", shadow = True, startangle = 180)
plt.show()
```

```python
#2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages. 
plt.axis("equal")
plt.title('Popularity of programming languages', pad = 60)
plt.pie(popularity, labels = languages, radius=1.5, autopct = "%0.1f%%", shadow = True, startangle = 180)
plt.show()
```

```python
#3. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.
#Make multiple wedges of the pie. 
plt.axis("equal")
plt.title('Popularity of programming languages', pad = 60)
explode = (0.1,0.1,0,0,0,0)
plt.pie(popularity, explode=explode, labels = languages, radius=1.5, autopct = "%0.1f%%", shadow = True, startangle = 180)
plt.show()
```

```python
#4. Write a Python programming to create a pie chart of gold medal achievements of five most successful countries 
#in 2016 Summer Olympics. 
#Read the data from a csv file. Sample data: 
#medal.csv 
#country,gold_medal 
#United States,46 
#Great Britain,27 
#China,26 
#Russia,19 
#Germany,17 
import pandas as pd
df = pd.read_csv('medal1.csv')
country_data = df["country"]
medal_data  = df["gold_medal"]
plt.title('Gold medal achievements of five most successful countries in 2016 Summer Olympics.', pad = 60)
explode = (0.1,0.1,0,0,0)
plt.pie(medal_data, labels = country_data, explode=explode , radius=1.5, autopct = "%0.1f%%", shadow = True, startangle = 180)
plt.show()
```
