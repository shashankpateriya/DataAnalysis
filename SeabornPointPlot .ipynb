```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Write a program to draw a point plot for sex against survived for a dataset given in
#url https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv
sns.set_theme(style="darkgrid")
titanic_dataset = sns.load_dataset("titanic")
sns.pointplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```
