```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Write a program to draw a swarm plot of total bill against size for a given dataset
#https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv
sns.set_theme(style="darkgrid")
tips_dataset = sns.load_dataset("tips")
sns.swarmplot(y="total_bill", x="size", data=tips_dataset, s = 3.5)
plt.show()
```

```python
#2. Write a program to draw swarm plot of total bill against day for a dataset given in
#url https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv
sns.set_theme(style="darkgrid")
tips_dataset = sns.load_dataset("tips")
fig = sns.swarmplot(y="total_bill", x="day", data=tips_dataset, s = 3.5)
plt.show()
```
