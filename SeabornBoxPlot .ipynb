```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Write a program to draw box plot of life expectancy by continent for a data set given in a url
#https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapmind er-FiveYearData.csv
sns.set_theme(style="darkgrid")
data_url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
gapminder = pd.read_csv(data_url)
sns.boxplot(y="lifeExp", x="continent", data=gapminder)
plt.show()
```

```python
#2. Write a program to draw a box plot of day by tips for a dataset given in a
#url https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv
sns.set_theme(style="darkgrid")
data_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
gapminder = pd.read_csv(data_url)
tips_dataset = sns.load_dataset("tips")
sns.boxplot(x="day", y="tip", data=tips_dataset)
plt.show()
```
