```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Write a program to draw a violin plot of sex against day for a given dataset
#https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv
sns.set_theme(style="darkgrid")
tips_dataset = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", hue="sex", split="True", data=tips_dataset)
plt.show()
```

```python
#2. Write a program to draw a violin plot of species against sepal length for a dataset given in a url 
#https://github.com/mwaskom/seaborn-data/blob/master/iris.csv
sns.set_theme(style="whitegrid")
iris_dataset = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=iris_dataset)
plt.show()
```
