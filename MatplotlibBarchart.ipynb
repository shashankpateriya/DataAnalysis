```python
#5. Write a Python programming to display a bar chart of the popularity of programming Languages.
#Attach a text label above each bar displaying its popularity (float value).
plt.figure(figsize = (9,7))
fig, ax = plt.subplots()    
width = 0.5
rects1 = ax.bar(x, popularity, width, color = ['black','red','orange','yellow','green','brown'])
plt.xlabel('Programming languages')
plt.ylabel('Popularity')
plt.title('Popularity of programming languages')
#for i, v in enumerate(popularity):
    #ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')
plt.ylim(0,25)
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
plt.show()
#plt.savefig(os.path.join('test.png'), dpi=300, format='png', bbox_inches='tight')
```

```python
ind = np.arange(len(x))

#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (9,7))
ax.bar(ind, popularity, width, color = ['black','red','orange','yellow','green','brown'])
#Now the trick is here.
#plt.text() , you need to give (x,y) location , where you want to put the numbers,
#So here index will give you x pos and data+1 will provide a little gap in y axis.
plt.ylim(0,25)
for index,data in enumerate(popularity):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=20), ha = 'center')
#plt.tight_layout()
plt.show()
```

```python
#10. Write a Python program to create bar plot of scores by group and gender.
#Use multiple X values on the same chart for men and women.
#Sample Data: 
#Means (men) = (22, 30, 35, 35, 26) 
#Means (women) = (25, 32, 30, 35, 29)
import numpy as np
game = ['G1','G2','G3','G4','G5']
men = [22, 30, 35, 35, 26] 
women = [25, 32, 30, 35, 29]
x_axis = np.arange(len(game))
plt.bar(x_axis - 0.2, men, 0.4, label = 'Men')
plt.bar(x_axis + 0.2, women, 0.4, label = 'Women')
plt.xticks(x_axis, game)
plt.xlabel("Men and women")
plt.ylabel("Scores")
plt.title("Scores by men and women")
plt.legend()
plt.show()
```

```python
#12. Write a Python program to create bar plots with error bars on the same figure.
#Sample Date Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824 
#Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

N = 4
ind = np.arange(N)
Mean = [0.2474, 0.1235, 0.1737, 0.1824]
SD = [0.3314, 0.2278, 0.2836, 0.2645]
fig = plt.subplots(figsize =(10, 7))
width = 0.5
p1 = plt.bar(ind, Mean, width, yerr = SD)
plt.xlabel('Model')
plt.ylabel('Standard deviation')
plt.show()
```

```python
#14. Write a Python program to create a stacked bar plot with error bars. 
#Note: Use bottom to stack the women?s bars on top of the men?s bars. 
#Sample Data: 
#Means (men) = (22, 30, 35, 35, 26) 
#Means (women) = (25, 32, 30, 35, 29) 
#Men Standard deviation = (4, 3, 4, 1, 5) 
#Women Standard deviation = (3, 5, 2, 3, 3) 
N = 5
 
men = (22, 30, 35, 35, 26)
women = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)  
width = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, women, width, yerr = womenStd)
p2 = plt.bar(ind, men, width,
             bottom = women, yerr = menStd) 
plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.legend((p1[0], p2[0]), ('women', 'men'))
plt.show() 
```

```python
#15. Write a Python program to create a horizontal bar chart with differently ordered colors.
#Note: Use bottom to stack the women?s bars on top of the men?s bars. 
#Sample Data Set: 
#languages = [['Language','Science','Math'], 
#['Science','Math','Language'], 
#['Math','Language','Science']] 
#numbers = [{'Language':75, 'Science':88, 'Math':96}, 
#{'Language':71, 'Science':95, 'Math':92}, 
#{'Language':75, 'Science':90, 'Math':89}] 
N = 3
students = ('S1','S2','S3')
languages = [['Language','Science','Math'], ['Science','Math','Language'], ['Math','Language','Science']] 
numbers = [{'Language':75, 'Science':88, 'Math':96}, {'Language':71, 'Science':95, 'Math':92},
           {'Language':75, 'Science':90, 'Math':89}]
ind = np.arange(N)
language = np.array([75,71,75])
science = np.array([88,95,90])
math = np.array([96,92,89])
fig = plt.subplots(figsize =(7, 5))
width = 0.35
p1 = plt.barh(ind, language, width, color='b')
p2 = plt.barh(ind, science, width, left=language, color='r')
p3 = plt.barh(ind, math, width, left=language+science, color='g')
plt.ylabel('People')
plt.title('Data of people')
plt.yticks(ind, students)
plt.legend((p1[0], p2[0], p3[0]), ('Language', 'Science','Math'))
plt.show() 
```

```python
#16. Write a Python program to create stack bar plot and add label to each section.
#Sample data: people = ('G1','G2','G3','G4','G5','G6','G7','G8') 
#segments = 4 

# multi-dimensional data 
#data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477], 
#[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847], 
#[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433], 
#[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]] 
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
N = 8
ind = np.arange(N)
x1 = np.array([ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477])
x2 = np.array([ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847])
x3 = np.array([ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433])
x4 = np.array([ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122])
#width = 0.35
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, x1, color='b')
p2 = plt.bar(ind, x2, bottom=x1, color='r')
p3 = plt.bar(ind, x3, bottom=x1+x2, color='g')
p4 = plt.bar(ind, x4, bottom=x1+x2+x3, color='black')
plt.ylabel('People')
plt.title('Data of people')
plt.xticks(ind, people)
plt.legend((p1[0], p2[0], p3[0], p4[0]), ('x1', 'x2','x3','x4'))
plt.show() 
```

```python
#17. Write a Python program to add textures (black and white) to bars and wedges.
#Note: Use bottom to stack the womens bars on top of the mens bars.
import matplotlib.pyplot as plt
fig = plt.figure()
patterns = ['//', '\\\\', '||', '--', '++', 'xx', 'oo', 'OO', '..', '**']
ax = fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i, 3, color='white', edgecolor='black', hatch=patterns[i])
plt.show()
```
